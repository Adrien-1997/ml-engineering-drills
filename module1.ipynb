## 1.1 Native data structures

### Lesson — list, dict, set, tuple

```python
# ── When to use what ───────────────────────────────────────
# list   → ordered sequence, index access, duplicates OK    O(1) append, O(n) search
# dict   → key→value, O(1) access, ordered insertion (3.7+)
# set    → uniqueness, O(1) membership test, no guaranteed order
# tuple  → immutable, hashable (usable as dict key)

# ── list ─────────────────────────────────────────────────────
scores = [85, 92, 78, 95, 88]
print("sorted      :", sorted(scores))                    # new list
print("sorted desc :", sorted(scores, reverse=True))
print("sort in-place:", scores)
scores.sort(reverse=True)
print("after sort() :", scores)

# sort with key
students = [("Alice", 85), ("Bob", 92), ("Charlie", 78)]
by_score = sorted(students, key=lambda x: x[1], reverse=True)
print("Ranking:", by_score)

# slicing
lst = list(range(10))
print("lst[2:5]  :", lst[2:5])
print("lst[::2]  :", lst[::2])    # step
print("lst[::-1] :", lst[::-1])   # reverse

# ── dict ─────────────────────────────────────────────────────
config = {"model": "lgbm", "n_estimators": 100, "lr": 0.05}

# safe access
print("get avec default:", config.get("max_depth", 6))

# merge (Python 3.9+)
overrides = {"n_estimators": 200, "max_depth": 7}
merged = config | overrides
print("merged:", merged)

# dict comprehension
squared = {x: x**2 for x in range(5)}
print("squared:", squared)

# iteration
for key, val in config.items():
    print(f"  {key}: {val}")

# ── set ──────────────────────────────────────────────────────
features_a = {"age", "income", "tenure", "charges"}
features_b = {"tenure", "charges", "nb_products", "region"}

print("intersection :", features_a & features_b)
print("union        :", features_a | features_b)
print("difference   :", features_a - features_b)   # in a, not in b
print("sym_diff     :", features_a ^ features_b)   # in one but not the other

# fast deduplication
raw = [1, 2, 2, 3, 3, 3, 4]
unique_ordered = list(dict.fromkeys(raw))   # preserves order (vs set)
print("deduplicated (order preserved):", unique_ordered)

# ── tuple ────────────────────────────────────────────────────
point = (3.0, 4.0)
x, y = point           # unpacking
print(f"point: x={x}, y={y}")

# tuple as dict key (hashable)
cache = {}
cache[(1, "train")] = 0.95
cache[(1, "test")]  = 0.91
print("cache:", cache)

# named tuple
from collections import namedtuple
ModelResult = namedtuple("ModelResult", ["name", "auc", "f1"])
r = ModelResult("lgbm", 0.94, 0.87)
print(f"{r.name}: AUC={r.auc}, F1={r.f1}")
```

### Lesson — collections: defaultdict, Counter, deque

```python
from collections import defaultdict, Counter, deque, OrderedDict

# ── defaultdict ───────────────────────────────────────────────
# Avoids KeyError on first insertion
word_lists = defaultdict(list)
for word in ["apple", "avocado", "banana", "blueberry", "cherry"]:
    word_lists[word[0]].append(word)
print("defaultdict(list):", dict(word_lists))

counts = defaultdict(int)
for c in "abracadabra":
    counts[c] += 1
print("defaultdict(int) :", dict(counts))

# ── Counter ───────────────────────────────────────────────────
tokens = "the quick brown fox jumps over the lazy dog the fox".split()
c = Counter(tokens)
print("Counter:", c)
print("most_common(3):", c.most_common(3))
print("total:", sum(c.values()))

# Counter operations
c2 = Counter({"the": 5, "fox": 1})
print("c + c2:", (c + c2)["the"])   # addition
print("c - c2:", (c - c2)["the"])   # subtraction (keeps > 0)

# ── deque ─────────────────────────────────────────────────────
# Double-ended queue : O(1) appendleft/popleft (vs list O(n))
dq = deque([1, 2, 3], maxlen=5)
dq.appendleft(0)    # O(1)
dq.append(4)        # O(1)
print("deque:", dq)

# Classic usage: sliding window with maxlen
window = deque(maxlen=3)
for val in [10, 20, 30, 40, 50]:
    window.append(val)
    print(f"  after {val}: {list(window)}")

# BFS queue
bfs_queue = deque(["start"])
visited = {"start"}
graph = {"start": ["A", "B"], "A": ["C"], "B": ["C"], "C": []}
order = []
while bfs_queue:
    node = bfs_queue.popleft()
    order.append(node)
    for neighbor in graph[node]:
        if neighbor not in visited:
            visited.add(neighbor)
            bfs_queue.append(neighbor)
print("BFS order:", order)
```

### Lesson — Comprehensions

```python
# ── List comprehensions ───────────────────────────────────────
numbers = list(range(20))

evens      = [x for x in numbers if x % 2 == 0]
even_sq    = [x**2 for x in numbers if x % 2 == 0]
nested     = [x*y for x in range(3) for y in range(3) if x != y]

print("evens   :", evens)
print("even_sq :", even_sq)
print("nested  :", nested)

# ── Dict comprehensions ───────────────────────────────────────
words = ["hello", "world", "python", "machine", "learning"]
lens  = {w: len(w) for w in words}
long_words = {w: l for w, l in lens.items() if l > 5}

print("word lengths :", lens)
print("long words   :", long_words)

# Invert a dict
original = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in original.items()}
print("inverted     :", inverted)

# ── Set comprehensions ────────────────────────────────────────
sentence = "the cat sat on the mat"
unique_chars = {c for c in sentence if c != " "}
print("unique chars :", sorted(unique_chars))

# Unique lengths
unique_lens = {len(w) for w in sentence.split()}
print("unique word lengths:", unique_lens)

# ── Generator expressions (lazy) ─────────────────────────────
import sys

list_comp = [x**2 for x in range(10_000)]
gen_expr  = (x**2 for x in range(10_000))

print(f"list size : {sys.getsizeof(list_comp):,} bytes")
print(f"gen size  : {sys.getsizeof(gen_expr):,} bytes   ← much lighter")

# sum over a generator — no need to materialize the list
total = sum(x**2 for x in range(10_000))
print(f"sum squares: {total:,}")

# ── Nested comprehension — flatten a list ───────────────
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat   = [x for row in matrix for x in row]
print("flat matrix:", flat)

# Transpose a matrix
transposed = [[row[i] for row in matrix] for i in range(3)]
print("transposed :", transposed)
```

### Lesson — *args, **kwargs, functions as objects

```python
from typing import Callable, Any
from functools import partial, reduce

# ── *args and **kwargs ─────────────────────────────────────────
def log_metrics(*metrics: float, model: str = "unknown", round_to: int = 4) -> None:
    "Accepts n positional values + named kwargs"
    formatted = [round(m, round_to) for m in metrics]
    print(f"[{model}] metrics: {formatted}")

log_metrics(0.94, 0.87, 0.91, model="lgbm")
log_metrics(0.88, model="logreg", round_to=2)

# Unpacking in the call
vals   = (0.91, 0.83)
kwargs = {"model": "rf", "round_to": 3}
log_metrics(*vals, **kwargs)

# ── Functions as objects ────────────────────────────────────
def square(x: float) -> float: return x ** 2
def cube(x: float) -> float:   return x ** 3

transforms = [square, cube, abs]
data = [-3, -1, 2, 4]

for fn in transforms:
    print(f"{fn.__name__:8s}: {[fn(x) for x in data]}")

# ── Higher-order function ────────────────────────────────
def apply_pipeline(data: list, *fns: Callable) -> list:
    "Applies a sequence of functions to each element"
    result = data
    for fn in fns:
        result = [fn(x) for x in result]
    return result

import math
result = apply_pipeline([1, 4, 9, 16], math.sqrt, lambda x: round(x, 2))
print("apply_pipeline:", result)

# ── functools.partial ─────────────────────────────────────────
def power(base: float, exp: float) -> float:
    return base ** exp

square_fn = partial(power, exp=2)
cube_fn   = partial(power, exp=3)

print("partial square:", [square_fn(x) for x in range(1, 6)])
print("partial cube  :", [cube_fn(x)   for x in range(1, 6)])

# ML use case: partial to fix hyperparams
import lightgbm as lgb
make_lgbm = partial(lgb.LGBMClassifier, n_estimators=200,
                    learning_rate=0.05, random_state=42, verbose=-1)
m1 = make_lgbm(num_leaves=31)
m2 = make_lgbm(num_leaves=63)
print(f"m1 num_leaves: {m1.num_leaves} | m2 num_leaves: {m2.num_leaves}")

# ── functools.reduce ──────────────────────────────────────────
product = reduce(lambda a, b: a * b, range(1, 6))   # 5! = 120
print(f"reduce (5!): {product}")
```

### Lesson — Closures

```python
# ── Closure: a function that captures its environment ──────
def make_multiplier(factor: float):
    "Returns a function that multiplies by factor"
    def multiply(x: float) -> float:
        return x * factor       # factor is captured (free variable)
    return multiply

double = make_multiplier(2)
triple = make_multiplier(3)

print("double(5):", double(5))   # 10
print("triple(5):", triple(5))   # 15
print("double.__closure__[0].cell_contents:", double.__closure__[0].cell_contents)

# ── Practical case: scaler factory ────────────────────────
def make_minmax_scaler(data: list):
    "Returns a min-max normalization function based on data"
    min_val = min(data)
    max_val = max(data)
    rng     = max_val - min_val

    def scale(x: float) -> float:
        return (x - min_val) / rng if rng != 0 else 0.0

    def unscale(x_scaled: float) -> float:
        return x_scaled * rng + min_val

    return scale, unscale

train_data = [10, 20, 30, 40, 50]
scale, unscale = make_minmax_scaler(train_data)

scaled   = [scale(x) for x in [10, 25, 50]]
unscaled = [unscale(x) for x in scaled]
print("scaled  :", scaled)
print("unscaled:", unscaled)

# ── Closure to store state (accumulator) ─────────────
def make_running_mean():
    "Accumulates values and returns the running mean"
    values = []

    def update(x: float) -> float:
        values.append(x)
        return sum(values) / len(values)

    return update

rm = make_running_mean()
for v in [10, 20, 30, 40]:
    print(f"  after {v}: mean = {rm(v):.2f}")

# ── Classic closure-in-loop trap ────────────────────
# ❌ Bad: all functions capture the same variable i
fns_bad = [lambda x: x * i for i in range(3)]
print("❌ fns_bad:", [f(2) for f in fns_bad])   # [4, 4, 4] — bug!

# ✅ Good: capture the current value with a default argument
fns_good = [lambda x, i=i: x * i for i in range(3)]
print("✅ fns_good:", [f(2) for f in fns_good])  # [0, 2, 4]
```

## 1.3 Generators and iterators

### Lesson — yield, generators, itertools

```python
from typing import Generator, Iterator, Iterable
from itertools import islice, chain, groupby, product, takewhile, dropwhile
import sys

# ── yield vs return ───────────────────────────────────────────
def count_up(start: int, stop: int) -> Generator[int, None, None]:
    "Generator: does not materialize the entire list"
    current = start
    while current < stop:
        yield current
        current += 1

gen = count_up(0, 5)
print("type :", type(gen))
print("next :", next(gen))   # 0
print("next :", next(gen))   # 1
print("list :", list(gen))   # [2, 3, 4] — resumes where we left off

# ── ML use case: batch data loader ─────────────────────────
import numpy as np

def batch_loader(X: np.ndarray, y: np.ndarray,
                 batch_size: int, shuffle: bool = True
                 ) -> Generator[tuple, None, None]:
    "Generates mini-batches (X_batch, y_batch) without loading everything"
    n = len(X)
    indices = np.random.permutation(n) if shuffle else np.arange(n)
    for start in range(0, n, batch_size):
        idx = indices[start:start + batch_size]
        yield X[idx], y[idx]

X = np.random.randn(100, 10)
y = np.random.randint(0, 2, 100)

for i, (X_b, y_b) in enumerate(batch_loader(X, y, batch_size=32)):
    print(f"batch {i}: X={X_b.shape}, y={y_b.shape}")

# ── Memory: generator vs list ────────────────────────────
list_mb  = sys.getsizeof([x**2 for x in range(100_000)])
gen_bytes = sys.getsizeof(x**2 for x in range(100_000))
print(f"list: {list_mb:,} bytes | generator: {gen_bytes} bytes")

# ── itertools ─────────────────────────────────────────────────
# islice: take the first N elements of an infinite iterable
def integers():
    n = 0
    while True:
        yield n; n += 1

first_10 = list(islice(integers(), 10))
print("islice:", first_10)

# chain: concatenate multiple iterables
a, b, c = [1, 2], [3, 4], [5, 6]
chained = list(chain(a, b, c))
print("chain:", chained)

# groupby: group consecutive elements by key
# IMPORTANT: data must be sorted by key before groupby
data = [("IDF", 100), ("IDF", 110), ("PACA", 80), ("PACA", 85), ("AURA", 120)]
for region, group in groupby(data, key=lambda x: x[0]):
    values = [v for _, v in group]
    print(f"  {region}: {values}")

# takewhile / dropwhile
nums = [1, 2, 3, 10, 4, 5]
taken   = list(takewhile(lambda x: x < 5, nums))
dropped = list(dropwhile(lambda x: x < 5, nums))
print("takewhile(<5):", taken)
print("dropwhile(<5):", dropped)
```

## 1.4 Decorators

### Lesson — writing a decorator from scratch

```python
import time
import functools
from typing import Callable, Any

# ── Decorator anatomy ──────────────────────────────────
def my_decorator(func: Callable) -> Callable:
    @functools.wraps(func)    # preserves __name__, __doc__
    def wrapper(*args, **kwargs):
        print(f"→ before {func.__name__}")
        result = func(*args, **kwargs)
        print(f"← after {func.__name__}")
        return result
    return wrapper

@my_decorator
def say_hello(name: str) -> str:
    return f"Hello, {name}!"

print(say_hello("Fabien"))
print("__name__ preserved:", say_hello.__name__)

# ── Useful decorator: timer ──────────────────────────────────
def timer(func: Callable) -> Callable:
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        t0 = time.perf_counter()
        result = func(*args, **kwargs)
        elapsed = time.perf_counter() - t0
        print(f"⏱ {func.__name__} took {elapsed*1000:.3f}ms")
        return result
    return wrapper

@timer
def train_model(n: int) -> float:
    "Simulates training"
    return sum(i**2 for i in range(n)) / n

train_model(100_000)

# ── Decorator with argument ──────────────────────────────────
def retry(max_attempts: int = 3, delay: float = 0.1):
    "Retries a function up to max_attempts times"
    def decorator(func: Callable) -> Callable:
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            last_exc = None
            for attempt in range(1, max_attempts + 1):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    last_exc = e
                    print(f"  attempt {attempt} failed: {e}")
                    time.sleep(delay)
            raise last_exc
        return wrapper
    return decorator

import random
random.seed(42)

@retry(max_attempts=3, delay=0.01)
def flaky_api_call() -> str:
    if random.random() < 0.6:
        raise ConnectionError("timeout")
    return "success"

try:
    result = flaky_api_call()
    print(f"flaky_api_call: {result}")
except ConnectionError as e:
    print(f"failed after retries: {e}")

# ── @property, @staticmethod, @classmethod ────────────────────
class ModelMetrics:
    def __init__(self, tp: int, fp: int, fn: int):
        self._tp, self._fp, self._fn = tp, fp, fn

    @property
    def precision(self) -> float:
        "TP / (TP + FP)"
        return self._tp / (self._tp + self._fp) if (self._tp + self._fp) > 0 else 0.0

    @property
    def recall(self) -> float:
        "TP / (TP + FN)"
        return self._tp / (self._tp + self._fn) if (self._tp + self._fn) > 0 else 0.0

    @property
    def f1(self) -> float:
        p, r = self.precision, self.recall
        return 2 * p * r / (p + r) if (p + r) > 0 else 0.0

    @staticmethod
    def from_arrays(y_true: list, y_pred: list) -> "ModelMetrics":
        "Factory method — computes TP/FP/FN from arrays"
        tp = sum(1 for t, p in zip(y_true, y_pred) if t == 1 and p == 1)
        fp = sum(1 for t, p in zip(y_true, y_pred) if t == 0 and p == 1)
        fn = sum(1 for t, p in zip(y_true, y_pred) if t == 1 and p == 0)
        return ModelMetrics(tp, fp, fn)

    @classmethod
    def perfect(cls) -> "ModelMetrics":
        return cls(tp=100, fp=0, fn=0)

m = ModelMetrics.from_arrays(
    y_true=[1, 1, 0, 1, 0, 0, 1],
    y_pred=[1, 0, 0, 1, 1, 0, 1]
)
print(f"precision={m.precision:.3f} recall={m.recall:.3f} f1={m.f1:.3f}")
print(f"perfect f1={ModelMetrics.perfect().f1:.3f}")
```

## 1.5 Error handling and context managers

### Lesson — try/except/finally + context managers

```python
import contextlib
import time
import logging
from pathlib import Path
from typing import Optional

# ── try / except / else / finally ────────────────────────────
def safe_divide(a: float, b: float) -> Optional[float]:
    try:
        result = a / b
    except ZeroDivisionError:
        print("  Error: division by zero")
        return None
    except TypeError as e:
        print(f"  Error: wrong type — {e}")
        return None
    else:
        # executed only if no exception
        print(f"  Success: {a}/{b} = {result:.4f}")
        return result
    finally:
        # always executed (cleanup)
        print("  finally block executed")

safe_divide(10, 3)
safe_divide(10, 0)

# ── Exception hierarchy — catch the right one ──────────────
class ModelError(Exception):
    "Base exception for ML errors"

class DataValidationError(ModelError):
    "Invalid input data"

class ModelNotFittedError(ModelError):
    "Model called without fit()"

def predict(model, X):
    if model is None:
        raise ModelNotFittedError("Call fit() before predict()")
    if X is None or len(X) == 0:
        raise DataValidationError("X cannot be empty")
    return [0.5] * len(X)

for case in [(None, [[1,2]]), ("model", []), ("model", [[1,2]])]:
    try:
        result = predict(*case)
        print(f"OK: {result}")
    except DataValidationError as e:
        print(f"DataValidationError: {e}")
    except ModelNotFittedError as e:
        print(f"ModelNotFittedError: {e}")
    except ModelError as e:
        print(f"ModelError (generic): {e}")

# ── Context manager avec __enter__ / __exit__ ─────────────────
class Timer:
    "Context manager to measure execution time"
    def __init__(self, name: str = ""):
        self.name    = name
        self.elapsed = 0.0

    def __enter__(self):
        self._start = time.perf_counter()
        return self   # available via `as`

    def __exit__(self, exc_type, exc_val, exc_tb):
        self.elapsed = time.perf_counter() - self._start
        print(f"⏱ [{self.name}] {self.elapsed*1000:.2f}ms")
        return False  # False = does not suppress the exception

with Timer("training") as t:
    total = sum(i**2 for i in range(100_000))
print(f"elapsed accessible: {t.elapsed*1000:.2f}ms")

# ── @contextmanager — more concise version ────────────────────
@contextlib.contextmanager
def managed_resource(name: str):
    "Simulates opening/closing a resource"
    print(f"→ opening {name}")
    resource = {"name": name, "data": []}
    try:
        yield resource
    except Exception as e:
        print(f"  error during {name}: {e}")
        raise
    finally:
        print(f"← closing {name}")

with managed_resource("BigQuery connection") as conn:
    conn["data"].append("row1")
    print(f"  using {conn['name']}: {conn['data']}")

# ── suppress: ignore a specific exception ───────────────
with contextlib.suppress(FileNotFoundError):
    Path("/nonexistent/file.txt").read_text()
print("continued after suppressed FileNotFoundError")
```
